```python
colleydf = pd.read_csv('/Users/ranbarasan/Documents/NFLWinLossNormalChart.csv')
```

```python
del colleydf['Unnamed: 0']
del colleydf['Unnamed: 33']
colleydf = colleydf.drop(index = 32)
```

```python
colleyModel = colleydf.to_numpy()
```

```python
for j in range(32):
    for i in range(32):
        if(i == j):
            colleyModel[i][j] = 16
        if(colleyModel[i][j] == 2):
            colleyModel[i][j] = -2 
        if(colleyModel[i][j] == 1):
            colleyModel[i][j] = -1
```

```python
colleyModel
```

```python
#colley formula for B
b = [6,
0,
-5,
3,
1,
0,
7,
-2,
-2.5,
4,
4,
5,
4,
4,
-6,
-3,
-1,
5,
1,
3,
-1,
0,
-2.5,
-1,
6,
0,
1,
-2,
4,
5,
-2,
-3]
```

```python
from scipy.linalg import solve
vec = solve(colleyModel,b)
vec
```

```python
sorted_indices = vec.argsort()
```

```python
sorted_indices
```

```python
ranked = [(teams[i], vec[i]) for i in sorted_indices]
```

```python
ranked.reverse()
```

```python
##SUCCESS: Colley Model
ranked
```
